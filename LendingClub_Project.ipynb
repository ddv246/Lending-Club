Identifying total number and percentage of missing values

```
def null_values(df):
        missing_val = df.isnull().sum()
        missing_val_percent = 100 * df.isnull().sum() / len(df)
        missing_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        missing_val_table_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        mis_val_table_columns = mis_val_table_columns[
            mis_val_table_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        print ("Dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_columns.shape[0]) +
              " columns that have missing values.")
        return mis_val_table_columns
```
